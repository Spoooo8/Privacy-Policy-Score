# Web Scrapping

```python
import requests
from bs4 import BeautifulSoup

website_url = input("Enter the website URL: ")

# Fetch the content of the website
response = requests.get(website_url)
soup = BeautifulSoup(response.content, 'html.parser')

privacy_policy_link = None

for link in soup.find_all('a', href=True):
    if 'privacy' in link.text.lower() or 'privacy' in link.get('href').lower():
        privacy_policy_link = link.get('href')
        break
         
if not privacy_policy_link:
    print("Privacy policy link not found on the given website.")
    

  
if not privacy_policy_link.startswith('http'):
    privacy_policy_link = website_url.rstrip('/') + '/' + privacy_policy_link.lstrip('/')

   
response = requests.get(privacy_policy_link)
soup = BeautifulSoup(response.content, 'html.parser')

   
privacy_policy_text = soup.get_text()

   
with open('privacy_policy.txt', 'w', encoding='utf-8') as file:
    file.write(privacy_policy_text)

print("Privacy policy saved")
```

# Types and Tokens

```python
import nltk
from nltk import word_tokenize, sent_tokenize

with open('privacy_policy.txt', 'r', encoding='utf-8') as file:
    text = file.read()
    
# Tokenize 
tokens = word_tokenize(text)

with open('tokenize.txt', 'w', encoding='utf-8') as file:
    for token in tokens:
        file.write(token + ',')

print("Tokens saved")
```

```python
#Types and token count
with open('tokenize.txt', 'r', encoding='utf-8') as file:
    tokens = file.read().split(',')

num_tokens = len(tokens)
num_types = len(set(tokens))

print(f"Number of tokens: {num_tokens}")
print(f"Number of types: {num_types}")
```

```python
# Importing libraries 
import spacy 
from wordcloud import WordCloud 
import matplotlib.pyplot as plt 

with open('privacy_policy.txt', 'r', encoding='utf-8') as file:
    tokens = file.read()


# Initializing the spaCy model instance 
nlp = spacy.load('en_core_web_sm') 

# Input text 
input_text = tokens


# Creating a spaCy document 
spacy_doc = nlp(input_text) 

# Initializing keywords list variable 
keywords = [] 

# Extracting keyphrases 
for chunk in spacy_doc.noun_chunks: 
	if chunk.text.lower() not in nlp.Defaults.stop_words: 
		keywords.append(chunk.text) 

# Displaying the keywords 
print(keywords) 


```

# K Mode Clustering

```python
pip install kmodes -q
```

```python
from kmodes.kmodes import KModes
import pandas as pd

tokens = keywords
df = pd.DataFrame(tokens, columns=['Token'])
data = df.values

k = 2

km = KModes(n_clusters=k, init='Huang', n_init=5, verbose=1)

clusters = km.fit_predict(data)

df['Cluster'] = clusters

df.to_csv('clustered_tokens.csv', index=False)

print("Clustering completed. Results saved to 'clustered_tokens.csv'.")
```

# Word Embedding

```python
from gensim.models import Word2Vec
import gensim
from nltk.tokenize import sent_tokenize, word_tokenize
import warnings
import requests
from bs4 import BeautifulSoup
import numpy as np
from sklearn.cluster import KMeans
```

```python
warnings.filterwarnings(action='ignore')

with open('privacy_policy.txt', 'r', encoding='utf-8') as file:
    input = file.read()

# Replaces escape character with space
input = input.replace("\n", " ")

data = []

# iterate through each sentence in the file
for i in sent_tokenize(input):
	temp = []
# tokenize the sentence into words
	for j in word_tokenize(i):
		temp.append(j.lower())

	data.append(temp)

# Create CBOW model
model = gensim.models.Word2Vec(data, min_count=1,
								vector_size=3, window=5)

# Print results
print("Cosine similarity between 'sell' " +
	"and 'data' - CBOW : ",
	model.wv.similarity('sell', 'data'))

print("Cosine similarity between 'protect' " +
	"and 'privacy' - CBOW : ",
	model.wv.similarity('protect', 'privacy'))

print("Cosine similarity between 'data' " +
	"and 'third-party' - CBOW : ",
	model.wv.similarity('data', 'third-party'))
```

```python
with open('word2vec.txt', 'w', encoding='utf-8') as f:
    for word in model.wv.index_to_key:
        vector = model.wv[word]
        vector_str = ' '.join(map(str, vector))
        f.write(f"{word} {vector_str}\n")
print("saved")
```

```python
with open('word2vec.txt', 'r', encoding='utf-8') as file:
    input_data = file.read()
    
# Initialize an empty list to store dictionaries
data_dicts = []

# Split input_data into lines
lines = input_data.strip().split('\n')

# Process each line
for line in lines:
    # Split line into items
    items = line.split()
    
    # Extract key
    key = items[0]  # First item is the key
    
    # Convert remaining items to float values
    values = list(map(float, items[1:]))
    
    # Create the dictionary
    data_dict = {key: values}
    
    # Append the dictionary to the list
    data_dicts.append(data_dict)

# Print all dictionaries
for idx, data_dict in enumerate(data_dicts, 1):
    print(data_dict)
```

# K Means Clustering

```python
data_str = input_data

# Parse the data
lines = data_str.strip().split('\n')
keys = []
values = []
for line in lines:
    parts = line.split()
    key = parts[0]
    value = list(map(float, parts[1:]))
    keys.append(key)
    values.append(value)

# Convert values to numpy array
values = np.array(values)

# Define number of clusters
k = 3 # You can choose an appropriate number of clusters

# Apply KMeans
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(values)
labels = kmeans.labels_

# Create a dictionary to map keys to clusters
clusters = {key: label for key, label in zip(keys, labels)}

# Print the resulting clusters
for key, cluster in clusters.items():
    print(f"'{key}' is in cluster {cluster}")
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

plt.figure(figsize=(10, 8))
scatter = plt.scatter(values[:, 0], values[:, 1], c=labels, cmap='viridis')

# Adding a color bar
colorbar = plt.colorbar(scatter)
colorbar.set_label('Cluster Label')

# Adding labels to axes
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('2D Scatter Plot of Clusters')

# Display the plot
plt.show()
```
